```python
import json
import pandas as pd

# # Display all rows and columns
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

# Load the data from the enterprise-attack.json file
with open('../data/enterprise-attack.json', 'r') as file:
    data = json.load(file)
```

```python
# Extract technique external_id and name values
techniques = []
for item in data['objects']:
    if item.get('type') == 'attack-pattern' and'external_references' in item and 'name' in item:
        for ref in item['external_references']:
            if ref.get('source_name') == 'mitre-attack' and 'external_id' in ref:
                techniques.append({'name': item['name'], 'external_id': ref['external_id']})

# Print the extracted values
for technique in techniques:
    print(f"technique_id: {technique['external_id']}, technique_name: {technique['name']}")
```

```python
# Create a DataFrame to display the information
techniques_df = pd.DataFrame(techniques).drop_duplicates()
techniques_df = techniques_df.sort_values(by=['external_id'], ascending=True)

#techniques_df.head() # display the first few rows to ensure it's correctly formatted
techniques_df
```
